```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.yello.ae/category/computer-services"
next_url="https://www.yello.ae/category/computer-services"
company_name=[]
company_link=[]
company_add=[]
for i in range(2,7):
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
    div=soup.find_all("div",class_="company_header")
    add_div=soup.find_all("div",class_="address")
    for i in div:
        company_name.append(i.find("h4").text)
        company_link.append("https://www.yello.ae/"+i.find("h4").find("a").get("href"))
    for i in add_div:
        company_add.append(i.text)
        # print(i.text)
    url=next_url+"/"+str(i)
```

```python
len(company_name)
```

```python
import pandas as pd
mydata=pd.DataFrame()
mydata["Company_name"]=company_name
mydata["Company_add"]=company_add
mydata["Company_link"]=company_link
```

```python
mydata.to_csv("yello's Data",index= False)
```
