```python
import requests
from bs4 import BeautifulSoup

# Function to scrape a single page
def scrape_page(page_num):
    url = f"https://www.yello.ae/category/computer-services={page_num}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Assuming each product is inside a div with class "product"
    products = soup.find_all('div', class_='product')
    
    for product in products:
        # Extracting product title and price as an example
        title = product.find('h2', class_='product-title').text
        price = product.find('span', class_='product-price').text
        print(f"Title: {title}, Price: {price}")

# Looping through multiple pages
for page in range(1, 6):  # Assuming there are 5 pages
    print(f"Scraping page {page}...")
    scrape_page(page)
```
